#### Variable Names
A variable can have a short name (like x and y) or a more descriptive name (age, carname, total_volume). Rules for Python variables:

###### 1. A variable name must start with a letter or the underscore character
###### 2. A variable name cannot start with a number
###### 3. A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
###### 4. Variable names are case-sensitive (age, Age and AGE are three different variables)

```python
# Assigning Values to Variables

vint = 50          # An integer assignment
vfloat  = 3.7       # A floating point
vstr   = "John"       # A string

# Multiple Assignment

a = b = c = 1
a,b,c = 1,2.8,"john"
```

### String 

Strings in Python are identified as a contiguous set of characters represented in the quotation marks.
The plus (+) sign is the string concatenation operator and the asterisk (*) is the repetition operator.

```python
str1 = 'Hello World!'

print(str1)         # Prints complete string
print(str1[0])       # Prints first character of the string
print(str1[2:5])     # Prints characters starting from 3rd to 5th
print(str1[2:])      # Prints string starting from 3rd character
print(str1 * 2 )     # Prints string two times
print(str1 + "TEST") # Prints concatenated string
```

## 6. For Loops

for `iterating variable` in `sequence` :
       
  - code block
    
   

```python
for x in range(10):
    #The range() function returns a sequence of numbers
    print(x)
```

###### for loop writting one line
- Single-Line For Loop

  for `iterating variable` in `sequence` : `command`

-  List Comprehension

   [`command` for `iterating variable` in `sequence`]

```python
for i in range(10): print(i)
    
#combine  with if statement
for i in range(10): print(i**2 if i<5 else 0)
    
# List Comprehension
print([i**2 for i in range(10)])
```
